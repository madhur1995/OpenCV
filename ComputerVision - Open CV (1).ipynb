```python
# step 1 : import haarcasede xml file
face_classifier = cv2.CascadeClassifier('.\Haarcascades\haarcascade_frontalface_default.xml')

# step 2 : import image
image = cv2.imread('.\Images\Modi.jpg')
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\grayscale.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\prabhas.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\dogsimage.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\sudoku.PNG"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\Modi.jpg"

# step 3 : convert RGB to Grey
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# step 4 :  creating ractangle on the faces

faces = face_classifier.detectMultiScale(gray, 1.3,5)

# step 5 : condition

if faces is ():
    print("No faces found")
    
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x,y),(x+w, y+h), (127,0,255),2)
    cv2.imshow('Face Detection', image)
    cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# step 1 : import haarcasede xml file
face_classifier = cv2.CascadeClassifier('.\Haarcascades\haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('.\Haarcascades\haarcascade_eye.xml')

# step 2 : import image
img = cv2.imread('.\Images\Trump.jpg')
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\grayscale.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\prabhas.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\dogsimage.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\sudoku.PNG"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\Modi.jpg"
# "C:\Users\Lenovo\Desktop\ComputerVision\Images\Trump.jpg"

# step 3 : convert RGB to Grey
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# step 4 :  creating ractangle on the faces

faces = face_classifier.detectMultiScale(gray, 1.3,5)

# step 5 : condition

if faces is ():
    print("No faces found")
    
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y),(x+w, y+h), (127,0,255),2)
    cv2.imshow('img', img)
    cv2.waitKey(0)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes=eye_classifier.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey),(ex+ew, ey+eh),(255,255,0),2)
        cv2.imshow('img', img)
        cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
def viewImage(image, name_of_window):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
```

```python
img = cv2.imread('opencv.jpg', 1)
viewImage(img, 'sample image')
```

```python
# cropping of image
cropped = img[100:1000, 100:1000]
viewImage(cropped, 'Image after cropping')
```

# How to capture video with computer webcam

```python
import cv2, time
video = cv2.VideoCapture(0)
check, frame = video.read()
time.sleep(1)
cv2.imshow('Capture', frame)
cv2.waitKey(0)
video.release()
cv2.destroyAllWindows()
```

```python
import cv2, time
video = cv2.VideoCapture(0)
a = 1
while True:
    a = a+1
    check, frame = video.read()    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GREY)
    cv2.imshow('Capture', gray)
    key = cv2.waitKey(1)
    
    if key == ord('q'):
        break      
print(f'No of frames captured : {a}')
video.release()
cv2.destroyAllWindows()
```
